```python
import split_folders
import sys
import os
from glob import glob
import pandas as pd
import numpy as np
from PIL import Image
```

```python
input_folder = ['breast_40', 'breast_100', 'breast_200', 'breast_400']
for f in input_folder:
    output_folder = f + '_split'
    # To only split into training and validation set, set a tuple to `ratio`, i.e, `(.8, .2)`.
    split_folders.ratio(f, output=output_folder, seed=1337, ratio=(.8, .1, .1)) # default values
    # Split val/test with a fixed number of items e.g. 100 for each set.
    # To only split into training and validation set, use a single number to `fixed`, i.e., `10`.
    split_folders.fixed(f, output=output_folder, seed=1337, fixed=(100, 100), oversample=False) # default values
```

```python
homeDir = 'C:/Users/PC/Jeehyun/ECE6780/Data/vgg/'
Mag_fact = ['breast_40_split', 'breast_100_split', 'breast_200_split', 'breast_400_split']
subfolder = ['train', 'test', 'val']
tums = ['benign','malignant']
ImgDirList = []
for M in Mag_fact:
    d1 = homeDir + M +'/'
    for subf in subfolder:
        d2 = d1 + subf +'/'
        for T in tums:
            d3 = d2 + T
            ImgDirList.append(d3)
```

```python
img_width, img_height = 224, 224
for f in ImgDirList:
    if f.split('/')[-1] == 'malignant':
        cls = 1
    else:
        cls = 0
    for filename in glob(f + '/*.png'):
        ImgData = Image.open(filename).convert('RGB').resize((img_width, img_height),Image.NEAREST)
        ImgData.save(filename, 'png')
```
